```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sea
```

```python
df2023 = pd.read_csv('date-hour-soo-dest-2023.csv.gz', names = ['date', 'hour', 'entry', 'exit', 'numRiders'])
df2024 = pd.read_csv('date-hour-soo-dest-2024.csv.gz', names = ['date', 'hour', 'entry', 'exit', 'numRiders'])
df2025 = pd.read_csv('date-hour-soo-dest-2025.csv.gz', names = ['date', 'hour', 'entry', 'exit', 'numRiders'])
```

```python
dfFull = pd.concat([df2023, df2024, df2025], axis = 0)
print(dfFull.shape)
dfFull.head()
```

```python
list(dfFull['entry'].unique())
```

```python
station_by_day = dfFull.groupby(['entry', 'date'])['numRiders'].sum().reset_index()
station_by_day
```

```python
# Create a reference series
reference_series = station_by_day[station_by_day['date'] == '2023-12-28'].set_index('entry')['numRiders']

# Map reference values to each row
station_by_day['refValue'] = station_by_day['entry'].map(reference_series)

# Calculate index (handle division by zero)
station_by_day['riderIndex'] = station_by_day.apply(lambda x: x['numRiders'] * 100 / x['refValue'], axis = 1)

# Drop reference value
station_by_day = station_by_day.drop('refValue', axis = 1)
```

```python
station_by_day
```

```python
plt.figure(figsize=(14, 8))
for station in ['WOAK']: #station_by_day['entry'].unique()[:1]:  # Plot first 10 stations
    station_data = station_by_day[station_by_day['entry'] == station]
    plt.plot(station_data['date'], station_data['riderIndex'], 
             label=station, marker='o', markersize=3)

plt.axhline(y=100, color='red', linestyle='--', alpha=0.7, label='Baseline (2023-12-28)')
plt.title('Rider Index Over Time by Station')
plt.xlabel('Date')
plt.ylabel('Rider Index (Normalized to 2023-12-28)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
